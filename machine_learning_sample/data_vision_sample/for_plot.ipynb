```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
a  = range(1, 16)
mean = 50
sigma = 10
b = np.random.normal(mean, sigma, 15).astype(int)
plt.plot(a, b)
#color
plt.plot(a, b, color='Red')
#line width
plt.plot(a, b, ls='--', lw=4)
plt.plot(a, b, marker='3', new=10)

#绘制直线

#绘制条形图
plt.bar(a, b)
#绘制饼状图
a = [3, 4, 5, 7, 12]
color_list = ['Red', 'Green', 'Blue', 'Blue', 'Yellow', 'Grey']
plt.pie(a, labels=['AA', 'BB', 'CC', 'DD', 'EE'], colors=color_list)

#绘制直方图
hist_data = np.random.normal(mean, 5, 500).astype(int)
plt.hist(hist_data)

#绘制散点图
scat1 = np.sort(np.random.normal(mean, 5, 50).astype(int))
scat2 = np.sort(np.random.normal(mean, 5, 50).astype(int))
plt.scatter(scat1, scat2)

#绘制箱线图
plt.boxplot(scat2)
```
